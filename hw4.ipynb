```python

corr = iris.corr(numeric_only=True)
print(corr)
```

```python

plt.figure(figsize=(8,6))
ham = sns.heatmap(corr, annot=True, cmap="Greens", center=0)
plt.title("corection iris")
plt.show()
```

```python
iris['petal_ratio'] = iris['petal_length'] / iris['petal_width']
iris.head()
```

```python
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle('Pairplotویژگی‌ها', y=1.02)
plt.show()
```

```python
# 3. Boxplot توزیع نسبت_گلبرگ بر اساس گونه
plt.figure(figsize=(10,6))
sns.boxplot (data=iris,x='species',y ='petal_ratio', palette='Set2')
plt.title('توزیع نسبت طول به عرض گلبرگ در سه گونه Iris')
plt.ylabel('نسبت طول گلبرگ به عرض گلبرگ')
plt.show()
```

```python
# 4. Scatter Plot: رابطه طول گلبرگ و نسبت
plt.figure(figsize=(10,6))
sns.scatterplot(data=iris, x="petal_length", y='petal_ratio',
                hue='species')
plt.title('correction ')
plt.xlabel('petal_length')
plt.ylabel('petal_ratio')
plt.legend()
plt.show()
```
