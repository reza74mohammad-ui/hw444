```python
import numpy as np
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
data = load_breast_cancer()
x = pd.DataFrame(data.data,columns=data.feature_names)
y = pd.Series(data.target)
y
```

```python
sample_b = x[y==1].iloc[:3].values
sample_b
```

```python
sample_m = x[y==0].iloc[:3].values
sample_m
```

```python
sample = np.vstack([sample_b,sample_m])
sample
```

```python
from scipy.spatial.distance import euclidean,cityblock,cosine,chebyshev,mahalanobis
cov = np.cov(x.values.T)
invcov = np.linalg.inv(cov)
```

```python
distance = []
for i in range(len(sample)):
      for j in range(i+1,len(sample)):
            p = sample[i]
            q = sample[j]
            distance_euclidean = euclidean(p,q) 
            distance_manhatan = cityblock(p,q)
            distance_cosine = cosine(p,q)
            distance_cheby = chebyshev(p,q)
            distance_maha = mahalanobis(p,q,invcov)
            distance.append({"اقلیدسی":distance_euclidean,"منهتن":distance_manhatan,"کسینوسی":distance_cosine,
                              "بزرگترین اختلاف":distance_cheby,"فاصله با توجه به پراکندگی":distance_maha})
distance_dataframe = pd.DataFrame(distance) 
print(distance_dataframe)                           
```
