```python
from sklearn.datasets import fetch_lfw_people
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data = fetch_lfw_people(min_faces_per_person=50,resize=0.4)
x = data.data
images = data.images
h,w = images[0].shape
w

```

```python
x.shape[0]
```

```python
x.shape[1]
```

```python
pca = PCA(n_components=4)
pca.fit(x)
```

```python
picture = pca.components_.reshape(4,h,w)
picture
```

```python
for i in range(4):
      plt.imshow(picture[i],cmap='gray')
      plt.show()
```
